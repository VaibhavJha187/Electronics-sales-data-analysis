### Importing Necessary Libraries


```python
import pandas as pd
import os
```

### Merging all months of sales data into a single file


```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data,df])
    
all_months_data.to_csv("all_data.csv", index = False)
```

### Read in updated dataframe

```python
all_data = pd.read_csv("all_data.csv")

all_data.head()
```

### Data cleaning !

##### Drop rows of NaN

```python
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()

all_data  = all_data.dropna(how ='all')
all_data.head()
```

#### Find 'Or' and delete it 

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

##### convert columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #convert to int

all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #convert to int
```

#### Augment data with additional columns

#### Task 1: Add column as Month

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Task 2 : Add a sales column 

```python
all_data['Total Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Task 4:  Add a city column

```python
# Let's use .apply()

def get_city(address):
    return address.split(',')[1]


def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+ ', '+ get_state(x))


all_data.head()
```

#### Question 1: What was the best month for sales ? How much was earned that month ?

```python
results = all_data.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,results['Total Sales'])

plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

#### Question 2: Which city had the highest number of sales?

```python
results = all_data.groupby('City').sum()

results
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities,results['Total Sales'])
plt.xticks(cities, rotation = 'vertical',size  = 8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City name')
plt.show()
```

#### Question 3: What time should they display advertisements to maximize likelihood of customer's buying product ?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data['Count'] = 1
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours,  all_data.groupby(['Hour']).count()['Order Date'])
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()

# My recommendation is around 10am to 11am or 6pm to 7pm
```

#### Question 4: What products are most often sold together ?

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df.loc[:, 'Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key, value in count.most_common(10):
    print(key,value)



```

```python
product_group = all_data.groupby('Product')

quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)

plt.ylabel('Quantity Ordered')
plt.xlabel('Product')

plt.xticks(products, rotation = 'vertical', size = 8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color = 'y')
ax2.plot(products, prices,'r-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color = 'g')
ax2.set_ylabel('Product Price ($)', color = 'b')
ax1.set_xticks(range(len(products)))
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)

plt.show()
```
